### 使用最简单的柱状图(Bar) 来分析2015 整年度的费用趋势

```python
# df_data_1.query('TRADE_YEAR == 2015')
group1 = df_data_1.query('TRADE_YEAR == 2015').groupby(['TRADE_YEAR','TRADE_MONTH'], as_index=False).sum()
# print(group1)

%brunel data('group1') bar x(TRADE_MONTH) y(AMOUNT) label(AMOUNT) style('.label{fill:white;text-shadow:none}') bar interaction(select:mouseover) color(#selection) + line x(TRADE_MONTH) y(AMOUNT)  effect(enter:2000)  title("2015年度 企业费用数据分析 V1")  axes(y:30000:grid:"金额",x:10:"月份"):: width=800, height=300
```

### 使用堆栈柱状图(Stack Bar) 来分部门查看

```python
group1 = df_data_1.query("TRADE_YEAR == 2015").groupby(['TRADE_YEAR','TRADE_MONTH','SEC_NAME'], as_index=False).sum()
# print(group1)
%brunel data('group1') bar x(TRADE_MONTH) y(AMOUNT) sum(AMOUNT) color(SEC_NAME) stack style('.label{fill:white;text-shadow:none}') bar interaction(select:mouseover)   effect(enter:2000)  title("2015年度 企业费用数据分析 V2")  axes(y:2000:grid:"金额",x:10:"月份"):: width=800, height=300
```

### 增加部门过滤条件，选取2个部门使用多列柱状图比较

```python
group1 = df_data_1.query("TRADE_YEAR == 2015 & (SEC_NAME == '人力资源部' | SEC_NAME == '生产管理部') ").groupby(['TRADE_YEAR','TRADE_MONTH','SEC_NAME'], as_index=False).sum()
# print(group1)
%brunel data('group1') bar x(TRADE_MONTH,SEC_NAME) y(AMOUNT) sum(AMOUNT) color(SEC_NAME) style('.label{fill:white;text-shadow:none}') bar interaction(select:mouseover)  color(SEC_NAME) effect(enter:2000)  title("2015年度 企业费用数据分析 V3")  axes(y:2000:grid:"金额",x:10:"月份"):: width=800, height=300
```

### 使用线图(Line) 来比较2014和2015的费用数据

```python
group1 = df_data_1.groupby(['TRADE_YEAR','TRADE_MONTH'], as_index=False).sum()
# print(group1)
%brunel data('group1') line x(TRADE_MONTH) y(AMOUNT) sum(AMOUNT) smooth(AMOUNT) color(TRADE_YEAR)  effect(enter:2000)  title("企业费用数据分析 V4 (2015 vs 2014)")  axes(y:3000:grid:"金额",x:12:"月份") legends(auto):: width=800, height=300
```

### 最简单的部门分析，也可用于文本的关键词展现

```python
group1 = df_data_1.query("TRADE_YEAR == 2015").groupby('SEC_NAME',as_index=False).sum()
# print(group1)
%brunel data('group1') cloud x(SEC_NAME) size(AMOUNT) color(SEC_NAME) sort(TRADE_MONTH) title('2015年度 企业费用数据分析 V5'):: width=800, height=300
```

### 使用泡泡图(Bubble) 直观的观察部门和省份的数据特征

```python
group1 = df_data_1.query("TRADE_YEAR == 2015").groupby(['SEC_NAME','LOCATION'],as_index=False).sum()
# print(group1)
%brunel data('group1') bubble color(SEC_NAME,LOCATION) size(AMOUNT) sum(AMOUNT) sort(SEC_NAME) label(LOCATION) tooltip(SEC_NAME, #all) title('2015年度 企业费用数据分析  V6') style('.label{font-size: 7pt;text-shadow:none}') interaction(panzoom) :: width=800, height=400
```

### 帅选出3月份员工的数据用弦图(Chord)来进一步观察

```python
group1 = df_data_1.query("TRADE_YEAR == 2015 & TRADE_MONTH == 3 & EMP_NO in ['A021','A022']").groupby(['SEC_NAME','LOCATION'],as_index=False).sum()
# print(group1)
%brunel data('group1') chord x(LOCATION) y(SEC_NAME) color(SEC_NAME) size(#count) sort(#count) label(LOCATION) tooltip(#all) title('2015年度 企业费用数据分析  V7') style('.label{font-size: 7pt;text-shadow:none}') interaction(panzoom) :: width=800, height=500
```

### 使用全国地图来标示费用热图(HeatMap)

```python
group1 = df_data_1.query("TRADE_YEAR == 2015").groupby(['LOCATION'], as_index=False).sum()

%brunel data('group1') map('china')  x(LOCATION) color(AMOUNT) label(LOCATION) title("2015年度 企业费用数据分析 V8") tooltip(AMOUNT,#all)  :: width=800, height=500
```

### 同时展现4个维度的数据：件数，月份，部门，金额

```python
group1 = df_data_1.query("TRADE_YEAR == 2015").groupby(['EMP_NO','TRADE_YEAR','TRADE_MONTH','SEC_NAME'], as_index=False).sum()
# print (group1)
%brunel data('group1') x(TRADE_MONTH) y(#count)  size(AMOUNT:500%) sum(AMOUNT) color(SEC_NAME) tooltip(AMOUNT,#all) title("2015年度 企业费用数据分析 V9","by IBM":footer) style('.footer{fill:red}') axes(x:"月份",y:"件数"):: width=800, height=300
```
